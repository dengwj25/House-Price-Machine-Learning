```python
import pandas as pd
import os
os.chdir('/Users/Washtu/Downloads/') 
os.getcwd()
pd.set_option('display.max_columns',100)
train_df = pd.read_csv('house_train.csv')
test_df=pd.read_csv('house_test.csv')
# Save the 'Id' column

train_ID = train_df['Unnamed: 0']
test_ID = test_df['Unnamed: 0']
# Now drop the 'Id' colum since we can not use it as a feature to train our model.
train_df.drop("Unnamed: 0", axis = 1, inplace = True)
test_df.drop("Unnamed: 0", axis = 1, inplace = True)
y_train = train_df['SalePrice']
X_train = train_df.drop('SalePrice', axis=1)
X_test = test_df.copy()
#Delete the dataframes that you do not need anymore to save memory.
del train_df, test_df
#Combine training and test dataframes before feature engineering.
```

```python
all_data = pd.concat([X_train, X_test], ignore_index=True)
all_data['TotalSF'] = all_data['TotalBsmtSF']+all_data['X1stFlrSF']+all_data['X2ndFlrSF']
all_data_1=all_data.drop(['TotalBsmtSF','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','X1stFlrSF','X2ndFlrSF'],axis=1)
all_data_1.head()
one_hot_df = pd.get_dummies(all_data_1, drop_first=True, dummy_na=True)
from sklearn.preprocessing import StandardScaler ##scaling the predictors
scaler = StandardScaler().fit(one_hot_df)
one_hot_df_scaled = pd.DataFrame(scaler.transform(one_hot_df), index=one_hot_df.index.values, columns=one_hot_df.columns.values)
train_new_1=one_hot_df_scaled.iloc[:1460, :]
train_new=pd.concat([y_train, train_new_1], axis=1)
test_new=one_hot_df_scaled.iloc[1460:, :]
train_new.head()
```
